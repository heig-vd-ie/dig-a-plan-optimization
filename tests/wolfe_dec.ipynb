```python
from pyomo.environ import *
```

```python


# Create a concrete model
model = ConcreteModel()

# Define variables
model.x = Var(domain=NonNegativeReals)
model.y = Var(domain=NonNegativeReals)

# Define constraints
model.c1 = Constraint(expr=3*model.x + 3*model.y <= 5)
model.c2 = Constraint(expr=model.x - 5*model.y <= 15)

extreme_points = set()

# Define a list of objective functions to try
objective_functions = [
    model.x,
    -model.x,
    model.y,
    -model.y,
    model.x + model.y,
    model.x - model.y,
    2*model.x + model.y,
    model.x + 2*model.y,
]

solver = SolverFactory('gurobi')  # You can use other LP solvers like 'cplex' or 'gurobi' if available


for obj_expr in objective_functions:
    model.objective = Objective(expr=obj_expr, sense=maximize)
    results = solver.solve(model)

    if results.solver.termination_condition == TerminationCondition.optimal:
        x_val = value(model.x)
        y_val = value(model.y)
        extreme_points.add((round(x_val, 6), round(y_val, 6))) # Rounding to handle potential floating-point issues

print("Dynamically found extreme points using Pyomo optimization:")
for point in extreme_points:
    print(f"({point[0]}, {point[1]})")
```
